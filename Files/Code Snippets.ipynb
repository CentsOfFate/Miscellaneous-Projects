# Seaborn Multiple Plots

## 2 Subplot Example

```python
# Load Pokemon Data
df = pd.read_csv('Pokemon.csv')

# Create Figure and Subplots
fig, ax = plt.subplots(1, 2, figsize=(15,8))

sns.regplot(x = "Attack", y = "Speed",
                   data = df,
                   ax = ax[0])

sns.regplot(x = "Sp. Atk", y = "Speed",
                   data = df,
                   ax = ax[1])

# Set Limits on X-Axis and Y-Axis so they are the same
ax[0].set_xlim(0, 200)
ax[1].set_xlim(0, 200)

ax[0].set_ylim(0, 200)
ax[1].set_ylim(0, 200)

# Write Titles for each
ax[0].set_title('Attack x Speed', fontsize = "x-large", fontweight = 'bold')
ax[1].set_title('Special Attack x Speed', fontsize = "x-large", fontweight = 'bold')

# Big Title
fig.suptitle("Example Plots - 2 Subplots", 
             fontsize = "xx-large", fontweight = 'bold')

plt.show()
```

## 3 Subplot Example

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("whitegrid")

# Load Pokemon Data
df = pd.read_csv('Pokemon.csv')

# Create a figure with Figure Size 10 x 10
fig = plt.figure(figsize = (15,15))

# Create Sub Plots
ax1 = fig.add_subplot(2, 2, 1) # Top Left Plot
ax2 = fig.add_subplot(2, 2, 2) # Top Right Plot
ax3 = fig.add_subplot(2, 1, 2) # Bottom Plot

# Top Left Plot
reg1 = sns.regplot(x = 'Attack', y = 'Defense', data = df,
                       ax = ax1)

# Top Right Plot
reg2 = sns.regplot(x = 'Attack', y = 'Sp. Atk', data = df,
           ax = ax2)

# Bottom Plot
box = sns.boxplot(x = 'Type 1', y = 'Attack', data = df,
            ax = ax3)

# Rotate Labels on X-Axis
box.set_xticklabels(box.get_xticklabels(), rotation = -45)


reg1.set_title('Regression Attack x Defense')
reg2.set_title('Regression Attack x Special Attack')
box.set_title('Box Plot of Pokemon Types and Attack')
fig.suptitle('Example Plot - 3 Subplots', 
             fontsize = "xx-large", fontweight = 'bold')
```
