# Example Seaborn Graphs (Fake Data)

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.pyplot as plt
import pyodbc
from matplotlib.ticker import FuncFormatter

sns.set_style('whitegrid')
```

```python
# Create Dataframe
df = pd.DataFrame({'Academic_Year' : ['2013-2014', '2014-2015', '2015-2016', '2016-2017', '2017-2018',
                                       '2013-2014', '2014-2015', '2015-2016', '2016-2017', '2017-2018'], 
                    'Proficiency' : [0.5124, 0.5234, 0.5465, 0.5342, 0.5621,
                                     0.5891, 0.6003, 0.5763, 0.5697, 0.5751],
                    'Type' : ['English Language Arts', 'English Language Arts', 'English Language Arts', 'English Language Arts', 'English Language Arts',
                              'Math', 'Math', 'Math', 'Math', 'Math']})

# Show Dataframe (Already Melted)
df
```

```python
# Set Style to White Grid
sns.set_style('whitegrid')

# Set Custom Color Palette
flatui = ["#DC4C46", "#3b86c6", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
sns.set_palette(flatui)

plt.figure(figsize = (15, 10), dpi = 300)
ax = sns.pointplot(x = "Academic_Year", y = "Proficiency", hue = "Type",
                   data = df)

# Title, x-axis, y-axis
plt.title('State Level Smarter Balanced 5 Year Trend', size = 'xx-large', fontweight = 'bold')
plt.xlabel('Academic Year', size = 'x-large', fontweight = 'bold')
plt.ylabel('Percent Proficient', size = 'x-large', fontweight = 'bold')

# Ticks larger
plt.xticks(size = 'large')
plt.yticks(size = 'large')

# Set Ticks
plt.ylim(0.50, 0.64)

# Set Y-Axis to be in Percentage Form
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))

# Set Legend
ax.legend(loc = 2, fontsize = 'x-large')

# Grab Percents for ELA and Math
elalist = df['Proficiency'].head(5)
mathlist = df['Proficiency'].tail(5)

# Label Data Points
[ax.text(p[0]-.2, p[1] +.005, round(p[1] * 100, 2), color = 'black', fontsize = 15) for p in zip(ax.get_xticks(), elalist)]
[ax.text(p[0], p[1] +.004, round(p[1] * 100, 2), color = 'black', fontsize = 15) for p in zip(ax.get_xticks(), mathlist)]

# Show Graph and Dataframe
plt.show()
```
