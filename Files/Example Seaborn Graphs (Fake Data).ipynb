# Example Seaborn Graphs (Fake Data)

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.pyplot as plt
import pyodbc
from matplotlib.ticker import FuncFormatter

sns.set_style('whitegrid')
```

```python
# Create Dataframe
df = pd.DataFrame({'Academic_Year' : ['2013-2014', '2014-2015', '2015-2016', '2016-2017', '2017-2018',
                                       '2013-2014', '2014-2015', '2015-2016', '2016-2017', '2017-2018'], 
                    'Proficiency' : [0.5124, 0.5234, 0.5465, 0.5342, 0.5621,
                                     0.5891, 0.6003, 0.5763, 0.5697, 0.5751],
                    'Type' : ['English Language Arts', 'English Language Arts', 'English Language Arts', 'English Language Arts', 'English Language Arts',
                              'Math', 'Math', 'Math', 'Math', 'Math']})

# Show Dataframe (Already Melted)
df
```

```python
# Set Style to White Grid
sns.set_style('whitegrid')

# Set Custom Color Palette
custom = ["#DC4C46", "#3b86c6", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
sns.set_palette(custom)


# Set Figure Size and DPI
fig = plt.figure(figsize = (19.20, 10.80), dpi = 300)

# Plot Points, set scale
ax = sns.pointplot(x = "Academic_Year", y = "Proficiency", hue = "Type",
                   data = df,
                   scale = 2)

# Title, X-Axis Label, Y-Axis Label
plt.title('State Level Smarter Balanced 5 Year Trend', size = 28, fontweight = 'bold')
plt.xlabel('Academic Year', size = 20, fontweight = 'bold')
plt.ylabel('Percent Proficient', size = 20, fontweight = 'bold')

# Ticks larger
plt.xticks(size = 18)
plt.yticks(size = 18)

# Set Ticks
plt.ylim(0.50, 0.64)

# Set Y-Axis to be in Percentage Form
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))

# Set Legend Location and Size
ax.legend(loc = 2, fontsize = 18)

# Grab Percents for ELA and Math
elalist = df['Proficiency'].head(5)
mathlist = df['Proficiency'].tail(5)

# Label Data Points
[ax.text(p[0] - .15, p[1] +.005, round(p[1] * 100, 2), color = 'black', fontsize = 18) for p in zip(ax.get_xticks(), elalist)]
[ax.text(p[0] - .004, p[1] +.006, round(p[1] * 100, 2), color = 'black', fontsize = 18) for p in zip(ax.get_xticks(), mathlist)]

# Show Graph and Dataframe
plt.show()

# Save 
#fig.savefig('State Level SBAC 5-Year (Presentation Style).png', dpi = 1200, bbox_inches = 'tight')
```

```python
# Create Dataframe
dff = pd.DataFrame({'Country' : ['USA', 'France', 'Spain', 'China', 'South Korea',
                                       'Germany', 'Canada', 'Norway'], 
                    'Revenue (In Thousands)' : [92, 63, 43, 101, 42,
                                                78, 33, 22]})

# Show Dataframe (Already Melted)
dff
```

```python
# Set Style to White Grid
sns.set_style('whitegrid')

# Set Custom Color Palette
custom = ["#DC4C46", "#3b86c6", "#20db2a", "#ef8fee", "#f4ef55",  "#fcc153"]
sns.set_palette(custom)

# Set Figure Size and DPI
fig = plt.figure(figsize = (19.20, 10.80), dpi = 300)

# Plot Points, set scale
ax = sns.barplot(x = "Country", y = "Revenue (In Thousands)",
                 data = dff,
                 palette = 'Set1')

# Title, X-Axis Label, Y-Axis Label
plt.title('Revenue by Country', size = 28, fontweight = 'bold')
plt.xlabel('Country', size = 20, fontweight = 'bold')
plt.ylabel('Revenue (In Thousands)', size = 20, fontweight = 'bold')

# Ticks larger
plt.xticks(size = 18)
plt.yticks(size = 18)

# Set Ticks
plt.ylim(0, 120)

# Label Data Points
[ax.text(p[0] - .1, p[1] + 2, p[1], color = 'black', fontsize = 18) for p in zip(ax.get_xticks(), dff['Revenue (In Thousands)'])]

# Show Graph and Dataframe
plt.show()
```

```python
import matplotlib.patches as mpatches

# Load Data
df = pd.read_csv('Full Pokemon.csv')

plt.figure(figsize = (15, 8), dpi = 100)

sns.distplot(df['Attack'], color = 'blue')
sns.distplot(df['Sp. Atk'], color = 'red')

blue_patch = mpatches.Patch(color = 'blue', label = 'Attack')
orange_patch = mpatches.Patch(color = 'red', label = 'Special Attack')
plt.legend(handles=[blue_patch, orange_patch], fontsize = 14)

# Title, X-Axis Label, Y-Axis Label
plt.title('Pokemon Attack/Special Attack Distribution Plot', size = 20, fontweight = 'bold')
plt.xlabel('Count Distribution', size = 16, fontweight = 'bold')
plt.ylabel('Proportion', size = 16, fontweight = 'bold')

# Ticks larger
plt.xticks(size = 12)
plt.yticks(size = 12)

# Set Ticks
plt.ylim(0, 0.020)

```

```python
# Load Pokemon Data
df = pd.read_csv('Full Pokemon.csv')

fdf = df.loc[df['Type 1'] == 'Fire']

# Create Figure and Subplots
fig, ax = plt.subplots(1, 2, figsize=(15,8))

sns.boxplot(x = "Type 1", y = "Attack",
                   data = fdf,
                   ax = ax[0],
                   color = 'orange')

sns.boxplot(x = "Type 1", y = "Sp. Atk",
                   data = fdf,
                   ax = ax[1],
                   color = 'red')

# Set Limits on Y-Axis so they are the same

ax[0].set_ylim(20, 180)
ax[1].set_ylim(20, 180)

# Write Titles for each
ax[0].set_title('Fire Types - Attack Box Plot', fontsize = "x-large", fontweight = 'bold')
ax[1].set_title('Fire Types - Special Attack Box Plot', fontsize = "x-large", fontweight = 'bold')

# Big Title
fig.suptitle("Example Plots - 2 Subplots", 
             fontsize = "xx-large", fontweight = 'bold')

# Set Y Labels
ax[0].set_ylabel('Attack Base Stat', fontsize = 14, fontweight = 'bold')
ax[1].set_ylabel('Special Attack Base Stat', fontsize = 14, fontweight = 'bold')

# Set X Labels
ax[0].set_xlabel('Type 1', fontsize = 14, fontweight = 'bold')
ax[1].set_xlabel('Type 1', fontsize = 14, fontweight = 'bold')

# Set Tick Size
ax[0].tick_params(labelsize = 12)
ax[1].tick_params(labelsize = 12)



plt.show()
```

```python
# Load Data
df = pd.read_csv('Full Pokemon.csv')

# Create a 2 x 2 Figure with Figsize (20, 15)
fig, ax = plt.subplots(ncols = 2, nrows = 2, figsize = (20, 15), dpi = 100)

y_df = ['Defense', 'Speed', 'Sp. Def', 'Sp. Atk']
y_vals = ['Defense', 'Speed', 'Special Defense', 'Special Attack']

# For Loop X-Axis, Y-Axis and Titles
for i in range(2):
    for j in range(2):
            sns.regplot(x = "Attack", y = y_df[i+i+j], data = df, ax = ax[i][j])
            ax[i][j].set_xlabel('Attack Base Stat', fontsize = 14, fontweight = 'bold')
            ax[i][j].set_ylabel(y_vals[i+i+j] + ' Base Stat', fontsize = 14, fontweight = 'bold')
            ax[i][j].set_title('Attack x ' + y_vals[i+i+j], fontsize = 18, fontweight = 'bold')
            ax[i][j].tick_params(labelsize = 12)
            ax[i][j].set_xlim(0, 200)
            ax[i][j].set_ylim(0, 250)

# Set Big Title
plt.suptitle('Pokemon Linear Regression Attack x Other Stats', fontsize = 24, fontweight = 'bold')
```
